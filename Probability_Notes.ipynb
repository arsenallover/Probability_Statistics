## Types of Data

1. Nominal : 
* Predetermined categories
* Cant be sorted
* Ex : Animal Class (mammal, reptile, fish)

2. Ordinal:
* Can be sorted
* Lacks scale
* Ex : Excellent, Good, Bad

3. Interval:
* provides scale
* no zero point
* Ex : Temperature

4. Ratio:
* Values have zero
* Ex : Length, weight

## Measures of Central Tendency

1. Mean : 
* Average Value
* Sensitive to outliers

2. Median :
* Middle Value

3. Mode :
* Frequent Value

They dont tell how spread the value is

## Measures of Dispersion

1. Range:
* Max - Min

2. Variance:
* Sum of sq distance from each point to Mean
* Sample Vs Population Variance: 
    * Sample variance needs bessel correction -> n-1
    * It makes it unbiased or rather doesnt underestimate the population variance.

3. Standard Deviation
* Square Root of variance
* Same unit as variable -> more interpretable

## Measures

1. Quartiles:
* Provides more information than simple mean aggregation
* Split the ordered data into 4 quarters
* Range -> max - min
* Median -> End of 2nd quarters/Begining of 3rd quarter

2. Inter Quartile Range (IQR):
* Difference btw start of 2nd and end of 3rd quarter
* Thats the box of box plot

3. Outlier
* Any data outside: +1.5*IQR, -1.5*IQR

## Covariance & Correlation Coefficient

1. Correlation Vs Causality
* Causality is cause and effect based
* Difficult to find Causality, need mroe stat analysis

2. Covariance:
* Compare the variances of each data in two features
* It tells are they spread out together and trending up and down together
* Normalize data -> before proceeding
* $ cov (x, y) = \frac{1}{N} \sum (x_i - \bar x)(y_i - \bar y) $

3. Correlation Coefficient:
* Covariance is not normalized. 
* Normalize Covariance to get Correlation Coefficient
* $ \rho (x, y) = \frac {cov(x, y)}{\sigma_x\sigma_y} $

* $ \rho (x, y) = \frac {\frac{1}{N} \sum (x_i - \bar x)(y_i - \bar y)} {\sqrt {\frac {\sum(x - \bar x)^2}{n}} \sqrt {\frac {\sum(y - \bar y)^2}{n}}} $

## Permutation Vs Combination

1. Permutation 
* Arrangement of objects in certain order
* Arrange a, b, c -> 3! = 6 ways -> abc, acb, bac, bca, cab, cba 

* Types:
    * Total Permutation : n!
    * Permutation of n taking r at given time : nPr = $\frac {n!}{n - r}! $
    * Permutation of n taking r at given time (with repetition) : $n^r! $

2. Combination

* unordered arrangements
* Selecting players in team (no matter the order) 
* Combination will always be lower than Permutation
* $nCr = \frac {n!}{r! (n - r)}! $

3. Permutation Vs Combination
* How many 3 letter combinations can be made from ABCDE
* Permutation -> 5P3 = 60
* Combination -> 5C3 = 10
* We collapse all the same orders in combinations

4. Combination with repetition
* Until now we observed no repetition
* Number of 3 letter combinations with ABCDE with repetition
* $ (n + r - 1)Cr = \frac {(n + r - 1)!}{r! (n - r)}! $
* Numerator becomes large -> allowing repetition

## Discrete Distributions

* Only discrete values (not values between them)
* aka probability mass functions
    * Uniform
    * poisson
    * Binomial

* Uniform Discrete Distributions
    * Rolling die - same probability

* Binomial Distributions
    * Two discrete mutually exclusive outcomes
    * H or T, ON or OFF, Success or Failure
    * They are called Bernoulli Trials (random experiment where only two possible outcomes)
    * Conditions of Binomial Distributions
        * Prob of Success, S is constant (with series of trails too) P(H) = 0.5
        * trials are independent
    * PMF : prob of 3 H in 10 trials, where H is heads prob
    * $ p(x : n, p) = (n \ C x)p^x (1 - p)^{n - x} $
    * $pmf = 10C3 * 0.5^3 * (1 - 0.5)^7 $

* Poisson Distributions
    * Similar to Binomial but takes success per unit time over trials
    * $\lambda  = \frac {num \ of \ occurances}{interval}$
    
    * $\lambda$ : No of cars seen in 5 mins, No. of coins seen in every mile

    * What is prob of seeing exactly 10 cars/mins

    * $P(x) = \frac{\lambda^x e^{-x}}{x!} $ 

    * Ex: Warehouse sees typically 8 deliveries a day. Whats prob of seeing only 4 tomorrow?

    * $P(x) = \frac{8^4 e^{-4}}{4!} = 0.057 $

    * Cumulative PMF : prob of seeing less than three -> p(0) + p(1) + p(2)

    * Scale lambda according to output. If lambda is given in hours and output wants in minutes -> scale them

## Continuous Distributions

* Normal Distributions
    * No bias to left or right
    * PDF - prob density function -> area under curve instead of summing all pmfs
    * $p(x) = \frac{1}{\sqrt{2\pi\sigma^2}}exp(-\frac{(x - \mu)^2}{2\sigma^2}) $
    * Standard Normal Distributions:
        * mean, stddev = 0, 1
        * 68, 95, 99 Rule
        * Standardize using z-scores
        * Using z-score, we can calculate xvalue percentile
        
        * $z = \frac{x - \mu}{\sigma} $

        * value from ztable maps the area to left of normal distribution

    * Consider A score 87 in test. The test had mean of 75, stddev of 7. Find how good is A?
        * zscore of A = (87 - 75)/7 = 1.7143
        * value of 1.7143 from ztable = 0.9564 -> 95.6 percentile
        * Implies area under the curve to left of A is 0.9564 -> A outscored 95.6 % of others
